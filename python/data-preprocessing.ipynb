```python
df_copy = df.copy()
df_copy.dropna(inplace = True)
df_copy.info()
```

```python
from sklearn.utils import resample

majority_class = df_copy[df_copy["class"] == "Positive"]
minority_class = df_copy[df_copy["class"] == "Negative"]

n_samples = len(minority_class)
majority_downsampled = resample(majority_class, replace = False, n_samples = n_samples, random_state = 42)

balanced_df = pd.concat([minority_class, majority_downsampled])
```

```python
balanced_df.groupby("class").size().reset_index(name = "count")
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
object_columns = df_copy.select_dtypes(include="object").columns.to_list()

for col in object_columns:
  df_copy[col] = le.fit_transform(df_copy[col])
```

```python
df_copy.info()
```

```python
X = df_copy.drop("class", axis = 1)
y = df_copy["class"]

smote = SMOTE(random_state = 42)
X_resampled, y_resampled = smote.fit_resample(X, y)

oversampled_df = pd.DataFrame(X_resampled, columns=X.columns)
oversampled_df["class"] = y_resampled
```
