```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
file_path = "datasets/last60.csv"
```

```python
df = pd.read_csv(file_path)

# Display the first 5 rows
print(df.head().to_markdown(index=False, numalign="left", stralign="left"))

# Print the column names and their data types
print(df.info())
```

```python
df.info()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Calculate and print the mean, standard deviation, and 25th, 50th (median), and 75th percentiles of `SuggestedRetail`
percentiles = np.percentile(df['SuggestedRetail'], [25, 50, 75])
print(f"Mean SuggestedRetail: ${df['SuggestedRetail'].mean().round(2)}")
print(f"Standard Deviation of SuggestedRetail: ${df['SuggestedRetail'].std().round(2)}")
print(f"25th Percentile of SuggestedRetail: ${percentiles[0].round(2)}")
print(f"50th Percentile of SuggestedRetail: ${percentiles[1].round(2)}")
print(f"75th Percentile of SuggestedRetail: ${percentiles[2].round(2)}")

# Plot a histogram of `SuggestedRetail`
plt.hist(df['SuggestedRetail'], bins=20, edgecolor='k')
plt.title('Distribution of Suggested Retail Prices')
plt.xlabel('Suggested Retail Price ($)')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
data.head()
```

```python

# Calculate descriptive statistics for the "SuggestedRetail" column
suggested_retail_stats = data['SuggestedRetail'].describe()

# Calculate skewness and kurtosis
suggested_retail_skewness = data['SuggestedRetail'].skew()
suggested_retail_kurtosis = data['SuggestedRetail'].kurtosis()

# Combine descriptive statistics, skewness, and kurtosis into a single DataFrame
summary_stats = pd.DataFrame({
    'Count': suggested_retail_stats['count'],
    'Mean': suggested_retail_stats['mean'],
    'Std Dev': suggested_retail_stats['std'],
    'Min': suggested_retail_stats['min'],
    '25%': suggested_retail_stats['25%'],
    '50%': suggested_retail_stats['50%'],
    '75%': suggested_retail_stats['75%'],
    'Max': suggested_retail_stats['max'],
    'Skewness': suggested_retail_skewness,
    'Kurtosis': suggested_retail_kurtosis
}, index=[0])  # Provide index explicitly

# Display the summary statistics
print(summary_stats)
```
